```python
from collections import deque

def BFS(a, b, target):
    # Map to mark visited states
    visited = {}
    is_solvable = False
    path = []

    # Queue for BFS
    q = deque()
    q.append((0, 0))  # Initial state: both jugs are empty

    while len(q) > 0:
        # Get the current state
        u = q.popleft()

        # If this state is already visited, skip it
        if (u[0], u[1]) in visited:
            continue

        # If any of the jugs has more water than its capacity, skip it
        if u[0] > a or u[1] > b or u[0] < 0 or u[1] < 0:
            continue

        # Add the current state to the path and mark as visited
        path.append([u[0], u[1]])
        visited[(u[0], u[1])] = True

        # If the target is reached, print the path
        if u[0] == target or u[1] == target:
            is_solvable = True
            if u[0] == target and u[1] != 0:
                path.append([u[0], 0])
            elif u[1] == target and u[0] != 0:
                path.append([0, u[1]])

            print("Path to solution:")
            for state in path:
                print(f"({state[0]}, {state[1]})")
            break

        # Pour water transitions
        q.append((u[0], b))  # Fill Jug2
        q.append((a, u[1]))  # Fill Jug1
        q.append((0, u[1]))  # Empty Jug1
        q.append((u[0], 0))  # Empty Jug2

        # Pour water from Jug1 to Jug2
        transfer = min(u[0], b - u[1])
        q.append((u[0] - transfer, u[1] + transfer))

        # Pour water from Jug2 to Jug1
        transfer = min(u[1], a - u[0])
        q.append((u[0] + transfer, u[1] - transfer))

    if not is_solvable:
        print("No solution")

# Test the function
Jug1, Jug2, target = 4, 3, 2
print("Path from initial state to solution state:")
BFS(Jug1, Jug2, target)
```
